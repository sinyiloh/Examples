# Sentiment Analysis

This exercise is to find out what demographic variables dictate 'Positive' or 'Not Positive' sentiment towards Sales Navigator based on LinkedIn posts globally over the last year?

Importing relevant Python packages for sentiment analysis. 

```python
import pandas as pd
import re
import string
import nltk
pd.set_option('display.max_colwidth', 100)

stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()

data = pd.read_csv("ga_global_posts.csv", encoding='ISO-8859-1')
```

Installing vaderSentiment to conduct sentiment analysis

!pip install vaderSentiment

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

Computing Sentiment Analysis Scores of the Posts

```python
data['sentiment_analysis'] = data['sharemediatext'].apply(lambda x: analyzer.polarity_scores(x))
```

Creating a new column with only the Compound Score

```python
def compound_extract(e):
    return e['compound']
```

```python
data['compound_score'] = data['sentiment_analysis'].apply(lambda x: compound_extract(x))
```

Generating the column that will contain my target variable: Positive or Not Positive

```python
def target_variable_fxn(a):
    if a > 0:
        return 'Positive'
    if a <= 0:
        return 'Not Positive'
```

```python
data['target_variable'] = data['compound_score'].apply(lambda x: target_variable_fxn(x))
```

Adding another column that adds a 1 for positive and a 0 for not positive

```python
def target_variable_fxn_value(a):
    if a > 0:
        return 1
    if a <= 0:
        return 0
```

```python
data['target_variable_value'] = data['compound_score'].apply(lambda x: target_variable_fxn_value(x))
```

```python
feature_cols = list(data_dummies)

feature_cols = [e for e in feature_cols if e not in ('id','sharemediatext','sentiment_analysis','compound_score','target_variable','target_variable_value')]

X = data_dummies[feature_cols]
y = data_dummies.target_variable_value

#from sklearn.preprocessing import StandardScaler

#scaler = StandardScaler()
#X_scaled = scaler.fit_transform(X)
```

```python
from imblearn.combine import SMOTETomek

from imblearn.combine import SMOTETomek

smt = SMOTETomek()
X_smt, y_smt = smt.fit_sample(X, y)
```

Smote-Tomek is applied. Now I need to check for a balanced data set. 

```python
X_smt_df = pd.DataFrame(X_smt)

y_smt_df = pd.DataFrame(y_smt)
```

Data set is now balanced. Each target variable type has 7K+ rows. Next step is to run logistic regression with k-fold validation. 

```python
from sklearn import metrics

import numpy as np

from sklearn import model_selection

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import f1_score

kf = model_selection.KFold(n_splits=5, shuffle=True)

f1_values = []
n = 0

print("~~~~ CROSS VALIDATION each fold ~~~~")
for train_index, test_index in kf.split(X_smt_df, y_smt_df):
    lr = LogisticRegression().fit(X_smt_df.iloc[train_index], y_smt_df.iloc[train_index].values.ravel()) 
    
    f1_values.append(metrics.f1_score(y_smt_df.iloc[test_index].values.ravel(), lr.predict(X_smt_df.iloc[test_index])))
    
    n += 1
    
    print('Model {}'.format(n))
    print('Avg. F1 Value: {}'.format(f1_values[n-1]))

print("~~~~ SUMMARY OF CROSS VALIDATION ~~~~")
print('Mean of F1 values for all folds: {}'.format(np.mean(f1_values)))
```
